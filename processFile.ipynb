```python
import pysolr
import csv, json
```

```python
SOLR_URL="http://localhost:8983/solr/searchProject/"
s = pysolr.Solr(SOLR_URL, timeout=10)
record_count = 0

import csv
with open('imdb.csv', 'r',encoding="utf8") as file:
    reader = csv.reader(file)
    for row in reader:
        record_count += 1
        if row[0]:
            genre = []
            genreList = row[3].split(";")
            for i in genreList:
                genre.append(i.strip())
            items=[{"id":record_count, "name":row[0], "date":row[1], "rate":row[2],"genre":genre}]
            s.add(items, commit=True)
            s.commit()
```
